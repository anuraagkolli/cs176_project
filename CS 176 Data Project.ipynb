```python
#Initialization
import pandas as pd 
import matplotlib.pyplot as plt


#Read Advanced.csv
full_df = pd.read_csv('Advanced.csv', sep = ',')
college_df = pd.read_csv('nba_college.csv', sep=',')
player_df = pd.read_csv('all_seasons.csv', sep=',')

#lg is league
#tm is team
#pos is position
display(full_df)
display(college_df)
display(player_df)
```

```python
#filter so that only specific columns are used
player_df = player_df[['player_name', 'team_abbreviation', 'age', 'player_height','player_weight', 'college', 'country', 'draft_year','draft_round', 'gp', 'pts', 'reb', 'ast']]

display(player_df)
```

```python
#Cleaning data

#drop missing values by deleting the column because it exists in another dataset
player_df = player_df.drop("college", axis=1)

#remove duplicates
duplicated = player_df.drop_duplicates()
```

```python
#Filtering data

#subset the deateset into various dataframes
usa_players = player_df[(player_df['country'] == "USA")]
display(usa_players)

#filter to get players that have more than 20 points per game
high_scorers = player_df[(player_df['pts'] > 20)]
display(high_scorers)
```

Indicates that 10,721 rows out of 12,844 rows are USA athletes.

Also shows that out of 12,844 rows, only 701 are top scorers with over 20 points per game.

```python
#Sorting data

#sort by decreasing order in age and increasing order by pts
age_pts_df = player_df.sort_values(by=['age','pts'], ascending=[False, True])
display(age_pts_df)

#sort by high scorers for average rebounds and assists
high_scorers_reb = high_scorers.sort_values(by=['reb', 'ast'], ascending = False)
display(high_scorers_reb)
```

The first sort helps to identify the players that are old yet high scorers. It also helps identify the players who are younger and still had a lot of points.

The second sort helps visualize who the top scorers mixed with the most rebounds and assists per game are. Essentially, telling us who valuable players on teams were (MVP).

```python
# subsetting Advanced.csv to wanted columns
full_df = full_df[['season', 'player_id', 'player', 'pos', 'age', 'experience', 'lg', 'tm', 'g', 'mp', 'per']]

# drop duplicates
full_df = full_df.drop_duplicates()

# change N/A to 0 because N/A values in the mp and per column could be set to 0
full_df["mp"].fillna(0, inplace=True)
full_df["per"].fillna(0, inplace=True)
full_df["age"].fillna(method="ffill", inplace=True)

display(full_df)
```

```python
#subset for only nba league players
nba_players = full_df[(full_df['lg'] == 'NBA')]
display(nba_players)

#subset for experience over 10
experienced_players = full_df[(full_df["experience"] > 10)]
display(experienced_players)
```

Shows that 29,413, which is majority of the rows in the dataset has athletes that are in the NBA.

Next subset highlights the experienced players in which only 3,117 rows have athletes that played for more than 10 years in the NBA

```python
# sort full_df by age in ascending order 
full_df = full_df.sort_values(by=['age'], ascending=False)
display(full_df)
```

Helps visualize the players that are older towards the top and the younger ones toward the bottom

```python
#clean, filter, and sort college_df

# define wanted columns 
college_df = college_df[['active_from', 'active_to','college', 'name', 'position','NCAA__3ptpg','NCAA_efgpct','NCAA_fgapg','NCAA_fgpct','NCAA_fgpg','NCAA_ft','NCAA_ftapg','NCAA_ftpg','NCAA_games','NCAA_ppg']]
# fill missing values with 0
college_df.fillna(0,inplace=True)

# drop duplicates
college_df.drop_duplicates()
```

```python
# sort in ascending order for years active 
college_df = college_df.sort_values(by=['active_from'], ascending=[True])
display(college_df)
```

Sorted the rows so that its in ascending order by when they were active from, noting the earliest players.

```python
# subset college_df for players with greater than 15 ppg
best_ncaa_players_df = college_df[(college_df['NCAA_ppg'] > 15)]
display(best_ncaa_players_df)
```

Subset created for NCAA players that scored greater than 15 points per game

```python
# scatter plot of weight and height
plt.figure(figsize=(8, 5))
plt.scatter(player_df['player_height'], player_df['player_weight'], color='green', marker="x")
plt.title('Player Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

The scatter plot above visualizes the relationship between player height and weight. Each point represents a player, with their height on the x-axis and weight on the y-axis. This plot can help in analyzing whether there's a correlation between a player's height and weight.

```python
# histogram of frequency of points per game
plt.figure(figsize=(8, 4))
plt.hist(player_df['pts'], bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Points Scored per Game')
plt.xlabel('Points per Game')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

The histogram above displays the distribution of points scored per game. Each bin represents a range of points, and the height of each bin indicates the frequency of players scoring within that range. 

```python
# Calculate average age by team
avg_age_by_team = player_df.groupby('team_abbreviation')['age'].mean()
avg_age_by_team.sort_values(inplace=True)

# plotting bar graph of average player age by team
plt.figure(figsize=(8, 4))
avg_age_by_team.plot(kind='bar', color='blue')
plt.title('Average Player Age by Team')
plt.xlabel('Team')
plt.ylabel('Average Age')
plt.xticks(rotation=65)
plt.show()
```

Here is the bar chart visualization showing the average player age by team. This chart is based on a simplified version of your data, as I created a fictional dataset to illustrate the concept. Each bar represents a different team, with the height of the bar indicating the average age of players in that team.

```python
# box plot to show correlation between draft round and points
plt.figure(figsize=(8, 5))
player_df.boxplot(by='draft_round', column=['pts'], grid=False)
plt.title('Points Per Game by Draft Round')
plt.xlabel('Draft Round')
plt.ylabel('Points Per Game')
plt.show()
```

The box plot above highlights the correlation between the draft round and the points per game they scored in the NBA. The plot allows us to see if the earliest drafted athletes have the most points per game compared to the rest. In fact, it does show that the first round picks tend to have higher points per game than second and third round picks, but on average, not by that much.

```python
# create a heat map to show correlations

corr_matrix_pre = player_df[["age", "player_height", "player_weight", "pts", "reb", "ast"]]
corr_matrix = corr_matrix_pre.corr()

plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()

labels = corr_matrix_pre.columns
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=60)
plt.yticks(ticks=range(len(labels)), labels=labels)

for i in range(len(labels)):
    for j in range(len(labels)):
        plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}")
```

We wanted to experiment with new types of plots we have not dealt with yet so we decided to explore the heat map. The heatmap above visualizes the correlation matrix for different attributes of players, including age, height, weight, points, rebounds, and assists per game.
